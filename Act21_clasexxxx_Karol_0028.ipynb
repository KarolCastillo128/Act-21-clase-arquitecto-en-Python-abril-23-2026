```python
class Arquitecto:
    def __init__(self, nombre, vision, resolucion, empatia):
        """
        Constructor de la clase.
        a) Los atributos definen el estado del objeto.
        """
        self.nombre = nombre
        self.vision_espacial = vision
        self.resolucion_problemas = resolucion
        self.empatia_social = empatia

    # b) Las funciones (métodos) definen las acciones
    def conceptualizacion_y_diseno(self, proyecto):
        print(f"[{self.nombre}]: Diseñando el concepto para '{proyecto}'.")
        print(f"   -> Aplicando: {self.vision_espacial}")

    def gestion_legal(self):
        print(f"[{self.nombre}]: Tramitando licencias ante el municipio.")
        print(f"   -> Estrategia: {self.resolucion_problemas}")

    def direccion_de_obra(self):
        print(f"[{self.nombre}]: Supervisando la construcción en sitio.")
        print(f"   -> Enfoque: {self.empatia_social}")

# --- Creación de Objetos (Instancias) ---

# Arquitecto 1: Especialista en Vivienda Moderna
arquitecto1 = Arquitecto(
    "Arq. Ana Valenzuela",
    "Modelado 3D de estructuras minimalistas",
    "Optimización de presupuestos limitados",
    "Diseño centrado en el confort del usuario"
)

# Arquitecto 2: Especialista en Restauración
arquitecto2 = Arquitecto(
    "Arq. Roberto Sierra",
    "Visualización de espacios históricos",
    "Refuerzo de estructuras antiguas",
    "Preservación del patrimonio cultural"
)

# --- Ejecución de Funciones ---

print("--- Actividades del Arquitecto 1 ---")
arquitecto1.conceptualizacion_y_diseno("Casa Bosque")
arquitecto1.direccion_de_obra()

print("\n--- Actividades del Arquitecto 2 ---")
arquitecto2.gestion_legal()
arquitecto2.conceptualizacion_y_diseno("Museo del Centro")
print("programa terminado")
print("clase arquitecto elaborada por Karol Castillo NL = 9")
```
